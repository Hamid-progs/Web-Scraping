# Scrap the Table!

```python
import pandas as pd

table = webpage.select('table.hockey-stats')[0]
columns = table.find_all('th')
columns_names = [c.string for c in columns]

table_rows = table.find('tbody').find_all('tr')
l = []
for tr in table_rows:
    td = tr.find_all('td')
    row = [str(tr.get_text()).strip() for tr in td] 
    l.append(row)
print(l)
```

```python
for L in l: 
    print(L)
```

```python
df = pd.DataFrame(l,columns=columns_names)
df.head()
```

```python
df.loc[df['Team'] != 'Did not play']
```

# Grab all the funfacts that use the word 'is'

```python
import re
li = webpage.select('ul.fun-facts li')
facts = [l.find(string=re.compile('is')) for l in li]
facts = [fact for fact in facts if fact]
facts
```

```python
facts = [fact.find_parent() for fact in facts if fact]
facts
```

```python
facts = [fact.find_parent().get_text() for fact in facts if fact]
facts
```

# Download the images

```python
url = 'https://keithgalli.github.io/web-scraping/'
r = requests.get(url+'webpage.html')
webpage = bs(r.content)

images = webpage.select('div.column img')
i=0
for image in images:
    image_url = images[i]['src']
    full_url = url + image_url

    image_data = requests.get(full_url).content
    with open(images[i]['alt']+'.jpg' , 'wb') as handler:
        handler.write(image_data)
    i = i+1
```

# Solve the Mystery challenge!

```python
files = webpage.select('div.block a')
relitive_files = [f['href'] for f in files]

url = 'https://keithgalli.github.io/web-scraping/'
for f in relitive_files:
    full_url = url + f
    page = requests.get(full_url)
    bs_page = bs(page.content)
    print(bs_page.find('p',attrs={'id':'secret-word'}).string)
```
